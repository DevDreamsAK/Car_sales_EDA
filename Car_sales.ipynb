# EDA

---



####1. Gender Distribution

```python
gd = sns.countplot(x='Gender',data=df, hue='Gender', palette='Set2', legend=False)
for bars in gd.containers:
  gd.bar_label(bars)
plt.title('Gender Distribution')
plt.show()
```

####2. Annual Income

```python
sns.histplot(df['Annual Income'], kde=True)
plt.title('Annual Income')
plt.show()
```

####3. Income Vs Price

```python
sns.scatterplot(x='Annual Income', y='Price ($)', data=df,hue='Gender')
plt.title('Annual Income Vs Price')
```

####4. Popular Car Companies

```python
df['Company'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Car Companies')
```

####5. Transmission Type

```python
tr = sns.countplot(x='Transmission',data=df, hue='Transmission', palette='mako', legend=False)
for bars in tr.containers:
  tr.bar_label(bars)
plt.title('Transmission Type')
plt.show()
```

####6. Body Type

```python
bd = sns.countplot(x='Body Style',data=df, hue='Body Style', palette='rocket', legend=False)
for bars in bd.containers:
  bd.bar_label(bars)
plt.title('Body Type')
```

####7. Price Distribution

```python
sns.histplot(df['Price ($)'], kde=True)
plt.title('Price Distribution')
plt.show()
```
